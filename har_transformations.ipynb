## Compara a base de Downstram com as bases de Pretexto

```python
from utils.plots import plot_instance_data
from data_modules.har import HarDataModule as HarDataModuleDownstram
from utils.enums import Sets, Datas
from data_modules.pretext import HarDataModule as HarDataModulePretext
from transforms.har import rotation, flip, noise_addition, permutation, scaling, time_warp, negation

# Carrega dados do HAR (Downstream)
data_module = HarDataModuleDownstram(batch_size=64)
data_loader = data_module.get_dataloader(set=Sets.TEST.value, shuffle=False)
it = iter(data_loader)
x, y = next(it)
x0, y0 = x[0], y[0]

# Carrega dados do HAR (MOTION)
transforms = [rotation, flip, noise_addition, permutation, scaling, time_warp, negation]
data_module = HarDataModulePretext(batch_size=64, main_data = Datas.MOTION)
data_loader = data_module.get_dataloader(set=Sets.TEST.value, shuffle=False, transforms=transforms)
it = iter(data_loader)
x, y = next(it)
x1, y1 = x[0], y[0]


# Carrega dados do HAR (MOBIT)
transforms = [rotation, flip, noise_addition, permutation, scaling, time_warp, negation]
data_module = HarDataModulePretext(batch_size=64, main_data = Datas.MOBIT)
data_loader = data_module.get_dataloader(set=Sets.TEST.value, shuffle=False, transforms=transforms)
it = iter(data_loader)
x, y = next(it)
x2, y2 = x[0], y[0]

# # Carrega dados do HAR (PAMAP)
# transforms = [rotation, flip, noise_addition, permutation, scaling, time_warp, negation]
# data_module = HarDataModulePretext(batch_size=64, main_data = Datas.PAMAP)
# data_loader = data_module.get_dataloader(set=Sets.TEST.value, shuffle=False, transforms=transforms)
# it = iter(data_loader)
# x, y = next(it)
# x3, y3 = x[0], y[0]

x = x1
y = y1

plot_instance_data(old_data=x0, oldTitle="Downstream", data=x1, title=Datas.MOTION.value)
plot_instance_data(old_data=x0, oldTitle="Downstream", data=x2, title=Datas.MOBIT.value)
# plot_instance_data(old_data=x0, oldTitle="Downstream", data=x3, title=Datas.PAMAP.value)
```
